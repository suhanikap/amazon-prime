# for comparing the top genres

```python
genres=['action','comedy','drama','horror']
counts=[50,80,60,40]
plt.figure(figsize=(8,5))
plt.bar(genres,counts,color='skyblue')
plt.xlabel("genres")
plt.ylabel("number of titles")
plt.title("Genres distribution in Amazon Prime")
plt.show()
```

# line chart

# trends over time

```python
years=[2010,2015,2020,2023]
titles_per_year=[150,300,500,700]
plt.figure(figsize=(8,5))
plt.plot(years,titles_per_year,marker='o',linestyle='-',color='purple')
plt.xlabel("Release year")
plt.ylabel("Number of titles")
plt.title("Amazon prime content Growth over years")
plt.grid()
plt.show()
```

```python
why choose line chart?
it shows continuous growth or decline
what are the insights?
easy to spot patterns
line chart connects data dots continuously
peaks in certain years could highlights blockbuster periods when major movies were released.
```

# histogram- distribution of numeric data

```python
runtimes=np.random.randint(60,180,100)
plt.figure(figsize=(8,5))
plt.hist(runtimes,bins=10,color='green',edgecolor='black')
plt.xlabel("Runtime(minutes)")
plt.ylabel("Frequency")
plt.title("Distribution of Movie Durations on Amazon prime")
plt.show()
```

```python
what insights we draw from it?
If the histogram shows a peak around 90–120 minutes, it confirms that most movies fall within the standard duration.
A long tail on the right suggests some exceptionally long films.
A left-skewed distribution (more low ratings) might indicate many poorly rated movies.
A right-skewed histogram (more high ratings) suggests Amazon Prime prioritizes highly-rated films.
A spike in recent years confirms Amazon has expanded content aggressive
If most shows have 1-2 seasons, it indicates Amazon favors shorter series.

```

# pie chart: Age certification distribution

```python
age_cert_counts=df['age_certification'].value_counts()
plt.figure(figsize=(7,7))
plt.pie(age_cert_counts,labels=age_cert_counts.index,autopct='%1.1f%%',colors=['lightblue','pink','lightgreen','orange'])
plt.title("Age Certification Distribution on Amazon Prime")
plt.show()
```

```python
what insights can we draw?
- largest section of the pie chart is 18+, Amazon Prime may prioritize adult content, catering to thrillers, dramas, and documentaries.
- Netflix often balances adult and teen content, while Disney+ leans heavily toward PG and family-friendly titles.
- If recent releases lean more toward PG-13, Amazon may be shifting focus to mainstream audiences, aiming for wider accessibility.
- If 18+ dominates, Amazon could double down on exclusive mature titles like high-budget crime dramas.
```

# Runtime analysis:Boxplot

```python
plt.figure(figsize=(7,6))
plt.boxplot(df['runtime'],vert=False,patch_artist=True)
plt.xlabel("Runtime(Minutes)")
plt.title("Distribution of Amazon Prime Movie Durations")
plt.show()
```

```python
what insihgts can be drawn?
- A higher median suggests Amazon prioritizes longer content, whereas a lower median suggests shorter films
- If plotted by release year, the box plot can highlight whether newer movies tend to be shorter or longer.
```
